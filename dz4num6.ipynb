```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000
u = np.matrix(np.random.normal(0, 1, n))
d = np.random.normal(0, 1, n)

A = u.T @ u / (u @ u.T) + np.diag(d)
A_1 = u.T @ u / (u @ u.T)
min_s = np.min(np.linalg.eigvalsh(A))
print(min_s)
print(np.linalg.eigvalsh(A_1))
```
