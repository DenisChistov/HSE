```python
import matplotlib.pyplot as plt
import numpy as np

a = np.zeros((15, 28))

a[2:-2 ,1] = 1; a[2, 2:6] = 1
a[2:7 ,6] = 1; a[7:-2, 7] = 1
a[7, 2:7] = 1; a[-3, 2:7] = 1
a[2:-2, 10] = 1; a[2:-2, 14] = 1
a[2:-2, 18] = 1; a[-3, 10:19] = 1
a[2, 21:-1] = 1; a[2:-2, -2] = 1
a[7, 21:-1] = 1; a[12, 21:-1] = 1

plt.imshow(a)
```

```python
u,m,v = np.linalg.svd(a)

for i in range(len(m)):
    if m[i] < 10e-99:
        print('rank(A) = ', i)
        rank = i
        break
        
h = len(u)
w = len(v)
```

```python
for k in range(rank+1):
    
    B = np.zeros((h, w))
    
    for i in range(k):
        B += m[i] * np.matrix((u.T)[i]).T * np.matrix((v)[i])
        
    plt.imshow(B)
    fig = plt.figure()
```

```python
print(*m, sep='\n')
```

```python
b = np.zeros((15, 28))

for i in range(rank):
    b[i,i] = m[i]

print(len(v))
print(len(b.dot(v)))
d = u.dot(b.dot(v))
plt.imshow(d);
```
